```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = df
```

```python
dropout_students = df1[df1['Target'] == 'Dropout']
graduate_students = df1[df1['Target'] == 'Graduate']
```

### คำนวณหาค่าเฉลี่ยของคะแนนสอบเข้าสำหรับนักเรียนในแต่ละสาขาวิชา โดยแบ่งเป็น กลุ่มนักเรียนที่ลาออก และ กลุ่มนักเรียนที่จบการศึกษา

```python
avg_admission_grade_dropout = dropout_students.groupby('Course')['Admission grade'].mean()
avg_admission_grade_graduate = graduate_students.groupby('Course')['Admission grade'].mean()
```

```python
print('Average admission grade dropout:',dropout_students['Admission grade'].mean())
```

```python
print('Average admission grade graduate:',graduate_students['Admission grade'].mean())
```

### คำนวณหาค่าเฉลี่ยของผลการเรียนในแต่ละเทอมสำหรับนักเรียนในแต่ละสาขาวิชา โดยแบ่งเป็น กลุ่มนักเรียนที่ลาออก และ กลุ่มนักเรียนที่จบการศึกษา

```python
mean_grades_dropout = dropout_students.groupby('Course')[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()
mean_grades_graduate = graduate_students.groupby('Course')[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()
```

```python
dropout_students[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()
```

```python
graduate_students[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()
```

###  คะแนนการสอบเข้าโดยเฉลี่ยของนักเรียนในแต่ละสาขาวิชา โดยแบ่งเป็น กลุ่มนักเรียนที่ลาออก และ กลุ่มนักเรียนที่จบการศึกษา

```python
print("Average admission grade for dropout students:", avg_admission_grade_dropout)
print('---------------------------------------------------------------------------------')
print("Average admission grade for graduate students:", avg_admission_grade_graduate)
```

### ผลการเรียนในแต่ละเทอมโดยเฉลี่ยของนักเรียนในแต่ละสาขาวิชา โดยแบ่งเป็น กลุ่มนักเรียนที่ลาออก และ กลุ่มนักเรียนที่จบการศึกษา

```python
print("Mean grades for dropout students by course:\n", mean_grades_dropout)
print('---------------------------------------------------------------------------------')
print("Mean grades for graduate students by course:\n", mean_grades_graduate)
```

```python
avg_admission_grades = pd.concat([avg_admission_grade_dropout, avg_admission_grade_graduate], axis=1)
avg_admission_grades.columns = ['Dropout', 'Graduate']
avg_admission_grades.reset_index(inplace=True)

plt.figure(figsize=(15, 8))
sns.barplot(y='Course', x='value', hue='variable', data=pd.melt(avg_admission_grades, ['Course']))
plt.title('Average Admission Grades by Course for Dropout and Graduate Students (0-200)')
plt.xlabel('Average Admission Grade (0-200)')
plt.ylabel('Course')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse_output=False)
course_onehot = onehot.fit_transform( df['Course'].values.reshape(-1, 1) )
onehot_cols_name = onehot.get_feature_names_out(['Course'])

onehot_df = pd.DataFrame(course_onehot, columns=onehot_cols_name)
onehot_df = pd.concat([onehot_df,  new_df.Target.reset_index().drop('index', axis=1) ], axis=1)


onehot_df
```

## นำตารางข้อมูลที่ถูก scale และตารางที่ผ่านการทำ onehot มาต่อกันเพื่อนำไปใช้เทรน

```python
train_df = pd.concat([scaled_df, onehot_df], axis=1)
```

```python
train_df.head()
```

## การแบ่งข้อมูล Train Test Split

```python
X = train_df.iloc[:, :-1]
y = train_df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=5)
# แบ่งข้อมูลโดยใช้เทคนิค Train Test Split โดยแบ่งสัดส่วนเป็นข้อมูลชุดฝึก 80% และข้อมูลทดสอบ 20% 
```

## การทำโมเดล

```python
knn = KNeighborsClassifier() # สร้างโมเดล KNN
```

## GridSearchCV
### ทำการหาพารามิเตอร์ที่ดีที่สุดในช่วงของข้อมูลที่กำหนดโดยใช้ GridSearchCV


```python
param_grid = {
    'n_neighbors': [x for x in range(1, 30)],
    'weights': ['uniform', 'distance'],
    }
clf = GridSearchCV(knn, param_grid, cv=10) ## ทำการ cross validation 10 ครั้ง
clf.fit(X_train, y_train)

print(clf.best_params_)
```

### ผลลัพธ์ที่ได้คือ ค่า n ที่ 22 และการให้ weight แบบ uniform ให้ผลลัพธ์ที่ดีที่สุด

## เทรนโมเดล KNN

```python
knn = KNeighborsClassifier( n_neighbors= clf.best_params_['n_neighbors'], weights=clf.best_params_['weights'])
knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
```

## Metric

```python
print('accuracy: ', accuracy_score(y_test, y_predict)) # ความแม่นยำของโมเดลเป็นเท่าไหร่
print('recall:   ', recall_score(y_test, y_predict)) # จากค่า graduate ทั้งหมดโมเดลทำนายได้อย่างถูกต้องเท่าไหร่
print('precision:', precision_score(y_test, y_predict)) # จากค่า prediction ของ graduate ทั้งหมดโมเดลทำนายทำนายได้อย่างถูกต้องเท่าไหร่
print('r2:       ', r2_score(y_test, y_predict)) # โมเดลอธิบายความแปรปรวนของข้อมูลได้เท่าไหร่
```

### โมเดลมีความแม่นยำที่ 88% และโมเดลสามารถอธิบายความแปรปรวนได้ที่ 52%

## Confusion Matrix

```python
cm = confusion_matrix(y_test, y_predict)
graph = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['dropout', 'graduate'])

graph.plot()
plt.show()
```

## วัดความ Overfitting / Underfitting

```python
y_predict = knn.predict(X_test)
print('accuracy on testing set: ', accuracy_score(y_test, y_predict))
y_predict = knn.predict(X_train)
print('accuracy on training set:', accuracy_score(y_train, y_predict)) # accuracy on training set
```

### คะแนนความแม่นยำของการทำนายจาก training set สูงกว่า testing set เล็กน้อย

## ทดสอบสมมุติฐานอัตราการว่างงานมีความสัมพันธ์เชิงเส้นกับการเรียนไม่จบหรือไม่ที่ระดับนัยสำคัญ 0.05

### H0 : อัตราการว่างงานไม่มีความสัมพันธ์เชิงเส้นกับการเรียนไม่จบ
### H1 : อัตราการว่างงานมีความสัมพันธ์เชิงเส้นกับการเรียนไม่จบ

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(df, x='Unemployment rate', kde=True, discrete=True, aspect=1.5)
plt.title("Unemployment rate distribution graph")
plt.show()
```

```python
from scipy import stats
```

```python
x = df['Unemployment rate']
y = df['Target']
rho, p_val = stats.spearmanr(x, y) # ข้อมูลเป็นประชากรจึงใช้ spearman
print(f'rho     : {rho}')    # ค่าสัมประสิทธิ์สหสัมพันธ์
print(f'p-value : {p_val}')  # ค่า p-value
```

### จากค่า p-value 0.61 ซึ่งมากกว่าค่านัยสำคัญ 0.05 จึงยอมรับ H0 นั่นคืออัตราการว่างงานไม่มีความสัมพันธ์เชิงเส้นกับการเรียนไม่จบเป็นจริง
